```python
import hashlib

class MD5HashTable:
    def __init__(self, m=65536):

        self.m = m  # Number of buckets (2^16 = 65536)
        self.table = [[] for _ in range(m)]  # Initialize hash table with empty lists

    def md5_hash_function(self, key):
        #Generate a hash for a string using MD5 and map it to a bucket index.
        md5_hash = hashlib.md5(key.encode()).hexdigest()  # Get MD5 hash as a hex string
        last_4_digits = md5_hash[-4:]  # Extract the last 4 hexadecimal digits
        return int(last_4_digits, 16) % self.m

    def insert(self, key):
        #Insert a string into the hash table, avoiding duplicates in the same bucket.
        index = self.md5_hash_function(key)
        if key not in self.table[index]:
            self.table[index].append(key)

# Function to handle file input and populate the hash table
def process_file(input_file_path):
    try:
        # Read words from the user-provided file
        with open(input_file_path, 'r') as file:
            words = file.read().splitlines()  # Read all lines and split into a list of words

        # Create an MD5-based hash table
        md5_ht = MD5HashTable()

        # Insert each word into the hash table and print its hash value
        for word in words:
            hash_value = md5_ht.md5_hash_function(word)
            md5_ht.insert(word)
            print(f"Word: {word}, Hash Value: {hash_value}")
    except FileNotFoundError:
        print(f"Error: File not found at path: {input_file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Main program
if __name__ == "__main__":
    input_file_path = input("Enter the path to the word.txt file: ")  # Prompt user for file path
    process_file(input_file_path)
```

```python
import random
import matplotlib.pyplot as plt

#Defination of  Universal hash function given in document
def universal_hash(x, a, p, m):
    return ((a * x) % p) % m


# Random hash function
def random_hash(x, m):
    return int(m * random.random())

# here we are Performing the experiment
def hash_experiment(m, num_trials=5):
 # Compare the behavior of a universal hash function and a random hash function.
    p = 524287  # Prime number
    results = {'universal': [], 'random': []}

    for _ in range(num_trials):
        # Universal hash function setup
        a = random.randint(1, p - 1)
        universal_buckets = [0] * m
        random_buckets = [0] * m

        # Hash all numbers from 1 to m
        for x in range(1, m + 1):
            # Universal hash
            u_index = universal_hash(x, a, p, m)
            universal_buckets[u_index] += 1

            # Random hash
            r_index = random_hash(x, m)
            random_buckets[r_index] += 1

        # program for Collecting statistics
        results['universal'].append((max(universal_buckets), min(universal_buckets)))
        results['random'].append((max(random_buckets), min(random_buckets)))

    return results

# Created Visualization function here
def visualize_results(results, m):
   #Visualize the results of the hash experiment using line plots.
    x = range(1, len(results['universal']) + 1)
    universal_max = [r[0] for r in results['universal']]
    universal_min = [r[1] for r in results['universal']]
    random_max = [r[0] for r in results['random']]
    random_min = [r[1] for r in results['random']]

    plt.figure(figsize=(12, 6))

    #here Plotting graph for Universal Hash Function
    plt.subplot(1, 2, 1)
    plt.plot(x, universal_max, label='Max Chain (Universal)', marker='o')
    plt.plot(x, universal_min, label='Min Chain (Universal)', marker='o')
    plt.title(f"Universal Hash Function (m={m})")
    plt.xlabel("Trial")
    plt.ylabel("Chain Length")
    plt.legend()

    #here Plotting graph for Random Hash Function
    plt.subplot(1, 2, 2)
    plt.plot(x, random_max, label='Max Chain (Random)', marker='o')
    plt.plot(x, random_min, label='Min Chain (Random)', marker='o')
    plt.title(f"Random Hash Function (m={m})")
    plt.xlabel("Trial")
    plt.ylabel("Chain Length")
    plt.legend()

    plt.tight_layout()
    plt.show()

# Main execution below
if __name__ == "__main__":
    m = 500000
    num_trials = 5
    results = hash_experiment(m, num_trials)
    visualize_results(results, m)

    # We are Printing the results in a table-like format
    print("Trial\tMax (Universal)\tMin (Universal)\tMax (Random)\tMin (Random)")
    for i in range(num_trials):
        universal_max, universal_min = results['universal'][i]
        random_max, random_min = results['random'][i]
        print(f"{i+1}\t{universal_max}\t\t{universal_min}\t\t{random_max}\t\t{random_min}")
```

```python
import hashlib
import random
import numpy as np

class UniversalHashFunction:
    def __init__(self, m, p=524287):
      # Initialize a universal hash function.
        self.m = m
        self.p = p
        self.a = random.randint(1, p-1)

    def hash(self, x):
        return ((self.a * x) % self.p) % self.m

def md5_last_4_digits_of_hexa_string(word):
  #Compute the last 4 hexadecimal digits of the MD5 hash for a given word.
    hash_object = hashlib.md5(word.encode())
    return int(hash_object.hexdigest()[-4:], 16)

def counting_trailing_zeroes(x):
  #Count the number of trailing zeroes in the binary representation of an integer
    if x == 0:
        return 0
    binary_representation = bin(x)[2:]
    return len(binary_representation) - len(binary_representation.rstrip('0'))

def flajolet_martin_algorithm(words, num_of_hashes=32):
  #Estimate the number of unique words using the Flajolet-Martin algorithm.
    m = 500000
    hashers = [UniversalHashFunction(m) for _ in range(num_of_hashes)]
    max_trailing_zeroes = [0] * num_of_hashes

    for word in words:
        id_ = md5_last_4_digits_of_hexa_string(word)
        for i in range(num_of_hashes):
            hash_value = hashers[i].hash(id_)
            trailing_zeroes = counting_trailing_zeroes(hash_value)
            if trailing_zeroes > max_trailing_zeroes[i]:
                max_trailing_zeroes[i] = trailing_zeroes

    #Here, we have Estimated the  number of unique words for each hash function
    estimates = [2 ** (z + 0.5) for z in max_trailing_zeroes]

    #We have  Returned the median estimate
    return np.median(estimates)

#program for reading the words from the file
with open('words.txt', 'r') as file:
    words = file.read().splitlines()

# Here we have Estimated the number of unique words
unique_words_estimate_of_file = flajolet_martin_algorithm(words)
print(f"Estimated number of unique words: {unique_words_estimate_of_file}")
```
