```python
#part(c) #calculate value of avg s1^2,avg s2^2 and avg s3^2
import random

D = [0.01 + (1000 - 0.01) * i / 99999 for i in range(100000)]

def compute_mean(sample):
    return sum(sample) / len(sample)

def compute_sample_variances(sample):
    n = len(sample)
    value_sample_mean = compute_mean(sample)
    
    s1_sq = sum((x - value_sample_mean) ** 2 for x in sample) / (n + 1)
    s2_sq = sum((x - value_sample_mean) ** 2 for x in sample) / n
    s3_sq = sum((x - value_sample_mean) ** 2 for x in sample) / (n - 1)
    
    return s1_sq, s2_sq, s3_sq

iter = 1000

#cumulative calculation
cumul_s1_sq = 0
cumul_s2_sq = 0
cumul_s3_sq = 0

avg_s1_sq_list = []
avg_s2_sq_list = []
avg_s3_sq_list = []

i = 1
while i <= iter:
    sample = [random.choice(D) for _ in range(50)]
    
    s1_sq, s2_sq, s3_sq = compute_sample_variances(sample)

    avg_s1_sq = cumul_s1_sq / i
    avg_s2_sq = cumul_s2_sq / i
    avg_s3_sq = cumul_s3_sq / i

    cumul_s1_sq += s1_sq
    cumul_s2_sq += s2_sq
    cumul_s3_sq += s3_sq
    
    avg_s1_sq_list.append(avg_s1_sq)
    avg_s2_sq_list.append(avg_s2_sq)
    avg_s3_sq_list.append(avg_s3_sq)
    
    i += 1

print(f"Avg value of s1^2 is : {avg_s1_sq}")
print(f"Avg value of s2^2 is : {avg_s2_sq}")
print(f"Avg value of s3^2 is : {avg_s3_sq}")
```

```python
#part(d)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(range(1, iter + 1), avg_s1_sq_list, label='Avg $s_1^2$')
plt.plot(range(1, iter + 1), avg_s2_sq_list, label='Avg $s_2^2$')
plt.plot(range(1, iter + 1), avg_s3_sq_list, label='Avg $s_3^2$')
plt.axhline(y=83333.167, color='r', linestyle='-', label='True Variance ($\\sigma^2$)')
plt.xlabel('Iter')
plt.ylabel('Variance')
plt.title('Convergence of Sample Variances to True Variance')
plt.legend()
plt.show()
```

```python
#part(e)
import random
import matplotlib.pyplot as plt

D = [0.01 + (1000 - 0.01) * i / 99999 for i in range(100000)]

def compute_mean(sample):
    return sum(sample) / len(sample)

def compute_sample_variances(sample):
    n = len(sample)
    sample_mean = compute_mean(sample)
    
    s1_sq = sum((x - sample_mean) ** 2 for x in sample) / (n + 1)
    s2_sq = sum((x - sample_mean) ** 2 for x in sample) / n
    s3_sq = sum((x - sample_mean) ** 2 for x in sample) / (n - 1)
    
    return s1_sq, s2_sq, s3_sq

#iterations
iter = 1000

repeats = 10

true_variance = 83333.167

all_avg_s1_sq = []
all_avg_s2_sq = []
all_avg_s3_sq = []

for _ in range(repeats):

    cumul_s1_sq = 0
    cumul_s2_sq = 0
    cumul_s3_sq = 0

    avg_s1_sq_list = []
    avg_s2_sq_list = []
    avg_s3_sq_list = []



    for i in range(1, iter + 1):
        sample = [random.choice(D) for _ in range(50)]
        
        s1_sq, s2_sq, s3_sq = compute_sample_variances(sample)
        
        cumul_s1_sq += s1_sq
        cumul_s2_sq += s2_sq
        cumul_s3_sq += s3_sq
        
        avg_s1_sq = cumul_s1_sq / i
        avg_s2_sq = cumul_s2_sq / i
        avg_s3_sq = cumul_s3_sq / i
        
        avg_s1_sq_list.append(avg_s1_sq)
        avg_s2_sq_list.append(avg_s2_sq)
        avg_s3_sq_list.append(avg_s3_sq)
    
    all_avg_s1_sq.append(avg_s1_sq_list)
    all_avg_s2_sq.append(avg_s2_sq_list)
    all_avg_s3_sq.append(avg_s3_sq_list)

# Function to plot results
def plot_results(avg_s1, avg_s2, avg_s3, true_variance, iter):
    plt.figure(figsize=(12, 6))
    plt.plot(range(1, iter + 1), avg_s1, label='Avg $s_1^2$')
    plt.plot(range(1, iter + 1), avg_s2, label='Avg $s_2^2$')
    plt.plot(range(1, iter + 1), avg_s3, label='Avg $s_3^2$')
    plt.axhline(y=true_variance, color='r', linestyle='-', label='True Variance ($\\sigma^2$)')
    plt.xlabel('Iterations')
    plt.ylabel('Variance')
    plt.title('Convergence of Sample Variances to True Variance')
    plt.legend()
    plt.show()

# Plot results for each repeat
for i in range(repeats):
    print(f"Plot for repeat {i + 1}")
    plot_results(all_avg_s1_sq[i], all_avg_s2_sq[i], all_avg_s3_sq[i], true_variance, iter)

#average convergence
avg_s1_converg = [abs(true_variance - avg_s1[-1]) for avg_s1 in all_avg_s1_sq]
avg_s2_converg = [abs(true_variance - avg_s2[-1]) for avg_s2 in all_avg_s2_sq]
avg_s3_converg = [abs(true_variance - avg_s3[-1]) for avg_s3 in all_avg_s3_sq]

print(f"Average absolute difference from true variance (s1^2): {sum(avg_s1_converg) / repeats}")
print(f"Average absolute difference from true variance (s2^2): {sum(avg_s2_converg) / repeats}")
print(f"Average absolute difference from true variance (s3^2): {sum(avg_s3_converg) / repeats}")


if (sum(avg_s1_converg) / repeats < sum(avg_s2_converg) / repeats) and (sum(avg_s1_converg) / repeats < sum(avg_s3_converg) / repeats):
    print("s1^2 converges to  true variance more quickly or frequently.")
elif (sum(avg_s2_converg) / repeats < sum(avg_s1_converg) / repeats) and (sum(avg_s2_converg) / repeats < sum(avg_s3_converg) / repeats):
    print("s2^2 converges to  true variance more quickly or frequently.")
else:
    print("s3^2 converges to  true variance more quickly or frequently.")


#reason why s3^2 converges to true variance more frequently
#s3^2  is designed to be an unbiased estimator of the population variance, which is why it converges more quickly
#and reliably to the true variance compared to s1^2 and s2^2.
#This property makes s3^2 the preferred estimator in many statistical applications, especially when working with smaller sample sizes
#where the bias in other estimators could have a more significant impact.
```
